```python
import gpxpy
import json

# Read the GPX file
with open('../../tests/data/running_zg_21252698158.gpx', 'r') as gpx_file:
	gpx = gpxpy.parse(gpx_file)

# Extract points and create the JSON structure
points = []
for track in gpx.tracks:
	for segment in track.segments:
		for point in segment.points:
			points.append({
				"latitude": {"value": point.latitude},
				"longitude": {"value": point.longitude}
			})

result = {"lap": {"points": points}}

# Write the JSON string to a file
with open('../../tests/data/grpcurl_CalculateLapDistance.json', 'w') as json_file:
	json_file.write(json.dumps(result, indent=2))
```
